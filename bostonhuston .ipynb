# Robust Regression(RANSAC)

```python
x=df['RM'].values.reshape(-1,1)
y=df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
rnc=RANSACRegressor()
```

```python
rnc.fit(x,y)
```

```python
inliner_mask=rnc.inlier_mask_
outliner_mask=np.logical_not(inliner_mask)
```

```python
line_x=np.arange(3,10,1)
line_y=rnc.predict(line_x.reshape(-1,1))
```

```python
sns.set(style='darkgrid',context="notebook")
plt.figure(figsize=(12,10))
plt.scatter(x[inliner_mask],y[inliner_mask],c='blue',marker='o',label='inliners')
plt.scatter(x[outliner_mask],y[outliner_mask],c='red',marker='s',label='outliners')
plt.plot(line_x,line_y,c='brown',label='ransac')
plt.show()
```

```python
rnc.estimator_.coef_
```

```python
rnc.estimator_.intercept_
```

# Evaluate

```python
from sklearn.model_selection import train_test_split
```

```python
x=df['LSTAT'].values.reshape(-1,1)
y=df['MEDV'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=True)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='k')
plt.xlim([-10, 50])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train,y_train_pred)
```

```python
mean_squared_error(y_test,y_test_pred)
```

# R2 score

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

# Near perfect model

```python
generate_random=np.random.RandomState(0)
x=10*generate_random.rand(1000)
y=3*x+np.random.randn(1000)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=True)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train.reshape(-1,1),y_train)
```

```python
y_train_pred=model.predict(x_train.reshape(-1,1))
y_test_pred=model.predict(x_test.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='k')
plt.xlim([-10, 50])
plt.show()
```

```python
mean_squared_error(y_train,y_train_pred)
```
